```python
data=data.drop(['S.No.'], axis=1)
data.info()
```

```python
import datetime as dt
current_year = dt.datetime.today().year
data['Car_Age'] = current_year - data['Year']
data.head()
```

```python
data['brand']=data.Name.str.split().str.get(0)
data['Model']=data.Name.str.split().str.get(1)+ data.Name.str.split().str.get(2)
data[['Name','brand','Model']]
```

```python
print(data.brand.unique())
print(data.brand.nunique())
```

```python
searchfor=['Honda','Audi','Maruti']
data[data.brand.str.contains('|'.join(searchfor))].head(5)
```

```python
data['brand'].replace({'ISUZU:Isuzu','Mini:Mini Cooper','Land:Land Rover'},inplace=True)
data.info()
```

```python
data.describe(include='all').T
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Recreate the 'brand' column
data['brand'] = data['Name'].str.split().str.get(0)

labels = data['brand'].value_counts().index
sizes = data['brand'].value_counts().values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Pie Chart of Brands')
plt.axis('equal')  
plt.show()
```

```python
import matplotlib.pyplot as plt

# Recreate the 'brand' column
data['brand'] = data['Name'].str.split().str.get(0)

labels = data['brand'].value_counts().index
sizes = data['brand'].value_counts().values

plt.figure(figsize=(12, 6))
plt.bar(labels, sizes, color=plt.cm.Paired.colors)
plt.title('Bar Chart of Brands')
plt.xlabel('Brands')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Recreate the 'brand' column
data['brand'] = data['Name'].str.split().str.get(0)

labels = data['Mileage'].value_counts().index
sizes = data['brand'].value_counts().values

plt.figure(figsize=(8, 8))
plt.hist(sizes, color='skyblue', edgecolor='black')
plt.title('Histogram of Brand Counts')
plt.xlabel('Number of Cars per Brand')
plt.ylabel('Frequency')
plt.show()
```
